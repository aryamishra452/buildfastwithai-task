```python
#IMPORTING LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#IMPORTING AND READING DATASET
data=pd.read_excel('/content/Data.xlsx')
df=pd.DataFrame(data)
df
```

## **STEP 1:CLEAN THE DATA**

```python
#Fix missing values (e.g., emails, LinkedIn URLs, job titles)
df['email'] = df['email'].fillna('Not Provided')
df['What is your LinkedIn profile?'] = df['What is your LinkedIn profile?'].fillna('Not Provided')
df['Job Title'] = df['Job Title'].fillna('Not Provided')
df
```

```python
#Normalize has_joined_event (Yes/No) values
df['has_joined_event'] = df['has_joined_event'].str.lower().map({
    'yes': 'Yes', 'y': 'Yes', 'true': 'Yes',
    'no': 'No', 'n': 'No', 'false': 'No'
})
df
```

```python
#Format amount, amount_tax, and amount_discount as numeric fields
cols_to_numeric = ['amount', 'amount_tax', 'amount_discount']
for col in cols_to_numeric:
    df[col] = pd.to_numeric(df[col].replace('[\$,]', '', regex=True), errors='coerce')
df
```

```python
#Extract and convert created_at into separate Date and Time columns
df['created_at'] = pd.to_datetime(df['created_at'], errors='coerce')
df['created_date'] = df['created_at'].dt.date
df['created_time'] = df['created_at'].dt.time
df.drop('created_at',axis=1,inplace=True)
df
```

```python
#Deliverable: Cleaned version of the data as a CSV or Excel file
df.to_excel("cleaned_data.xlsx", index=False)
from google.colab import files
files.download("cleaned_data.xlsx")
```

## **STEP 2:ANALYZE KEY METRICS**

```python
#1.Conversion Funnel Metrics

#Total number of approved users
total_approved = df[df['approval_status'].str.lower() == 'approved']
total_approved_count = len(total_approved)

#Number and % who joined the event
joined = total_approved[total_approved['has_joined_event'] == 'Yes']
joined_count = len(joined)
joined_percent = (joined_count / total_approved_count) * 100

#Show-up vs no-show rate
no_show = total_approved[total_approved['has_joined_event'] == 'No']
no_show_count = len(no_show)
no_show_percent = (no_show_count / total_approved_count) * 100

print(f"Total Approved: {total_approved_count}")
print(f"Joined: {joined_count} ({joined_percent:.2f}%)")
print(f"Show: {total_approved_count-no_show_count} ({(100-no_show_percent):.2f}%)")
print(f"No Show: {no_show_count} ({no_show_percent:.2f}%)")
```

```python
# Joined vs Did not join chart
labels = ['Joined', 'Did Not Join']
counts = [joined_count, (total_approved_count-joined_count)]

plt.figure(figsize=(6, 4))
plt.bar(labels, counts, color=['#4CAF50', '#F44336'])
plt.title('Number of Approved Users Who Joined the Event')
plt.ylabel('Number of Users')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Show-up vs no-show-rate chart
sizes = [joined_count, no_show_count]
labels = ['Showed Up', 'No-show']
colors = ['#2196F3', '#FFC107']

plt.figure(figsize=(5, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Show-up vs No-show Rate')
plt.axis('equal')  # Equal aspect ratio ensures circle
plt.show()
```

```python
# Top 5 most common job titles
top_jobs = df['Job Title'].value_counts().head(5)

# % ofStudent vs Professional
job_category_percent=df['Job Title'].str.lower().apply(
    lambda x: 'Student' if 'student' in str(x) else 'Professional').value_counts(normalize=True) * 100

# Duplicate/suspicious entries
duplicate_emails = df[df.duplicated('email', keep=False)]
duplicate_linkedin = df[df.duplicated('What is your LinkedIn profile?', keep=False)]
missing_names = df[df['first_name'].isna() | df['last_name'].isna()]

suspicious_entries = pd.concat([duplicate_emails, duplicate_linkedin, missing_names])
suspicious_entries = suspicious_entries.drop_duplicates()

print("Top 5 job titles:")
print(top_jobs)
print("\nStudent vs Professional:")
print(job_category_percent)
print("\nDuplicate/suspicious entries:")
print(suspicious_entries)
```

```python
# % ofStudent vs Professional chart

def classify_job(title):
    title = str(title).lower()
    if 'student' in title:
        return 'Student'
    return 'Professional'

df['job_category'] = df['Job Title'].apply(classify_job)
job_counts = df['job_category'].value_counts()

# Plot
plt.figure(figsize=(5, 5))
plt.pie(job_counts, labels=job_counts.index, autopct='%1.1f%%', colors=['#00BCD4', '#9C27B0'])
plt.title('% of Students vs Professionals')
plt.axis('equal')
plt.show()
```

```python
df['What is your LinkedIn profile?'] = df['What is your LinkedIn profile?'].fillna('')

valid_linkedin = df['What is your LinkedIn profile?'].str.startswith('https://www.linkedin.com/in/')
valid_count = valid_linkedin.sum()
invalid_count = (~valid_linkedin).sum()
print(f"Valid LinkedIn URLs: {valid_count}")
print(f"Invalid LinkedIn URLs: {invalid_count}")
```

### **STEP 3: ADDITIONAL EXPLORATION**

```python
# Sort by 'created_date' and 'created_time' together without creating a new column
df = df.sort_values(by=['created_date', 'created_time']).reset_index(drop=True)

# Calculate cutoff index (20% mark)
cutoff = int(len(df) * 0.2)

# Label top 50% as Early, bottom 50% as Late
df['registration_segment'] = ['Early' if i < cutoff else 'Late' for i in df.index]
print(df['registration_segment'].value_counts())
```
